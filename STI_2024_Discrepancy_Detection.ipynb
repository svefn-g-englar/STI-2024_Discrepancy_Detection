```python
punctuations = ["।", ",", ";", ":", "?", "!", "'", ".", "\"", "-",
                "[", "]", "{", "}", "(", ")", '–', "—", "―", "~"]

punc = ''.join(punctuations)
punc
```

```python
translator = Translator()
bnorm = Normalizer(allow_english=True)

# Regex for matching zero witdh joiner variations.
STANDARDIZE_ZW = re.compile(r'(?<=\u09b0)[\u200c\u200d]+(?=\u09cd\u09af)')

# Regex for removing standardized zero width joiner, except in edge cases.
DELETE_ZW = re.compile(r'(?<!\u09b0)[\u200c\u200d](?!\u09cd\u09af)')

# Regex matching punctuations to remove.
PUNC = re.compile(r'[`£|¢|Ñ+*/=–•।!()\[\]{};:%\',"“\\’<>./?@#$^।&*_~‘—॥”‰�￰￰।,;:?!\'.\"-[\]{}()–—―~]+')

def remove_hashtags(text):
    pattern = re.compile(r'#[\u0980-\u09FF\w]+')
    result = pattern.sub('', text)

    return result

def remove_digits(text):
    pattern = re.compile(r'\d+|[\u09E6-\u09EF]+')
    result = pattern.sub('', text)

    return result


def removeOptionalZW(text):
    """
    Removes all optional occurrences of ZWNJ or ZWJ from Bangla text.
    """
    text = STANDARDIZE_ZW.sub('\u200D', text)
    text = re.sub(r'\u200d', '', text)
    text = DELETE_ZW.sub('', text)
    return text

def removePunc(text):
    """
    Remove for punctuations from text.
    """
    text = PUNC.sub(r" ", text)
    return text

def normalizeUnicode(text, normalize_nukta=True):
    """
    Normalizes unicode strings using the Normalization Form Canonical
    Composition (NFC) scheme where we first decompose all characters and then
    re-compose combining sequences in a specific order as defined by the
    standard in unicodedata module. Finally all zero-width joiners are
    removed.
    """
    if normalize_nukta:
        words = [ bnorm(word)['normalized']  for word in text.split() ]
        text = " ".join([word for word in words if word is not None])
        text = text.replace("\u2047", "-")

    text = text.replace(u"\u098c", u"\u09ef")
    text = unicodedata.normalize("NFC", text)
    text = removeOptionalZW(text)
    text = remove_hashtags(text)
    text = remove_digits(text)
    text = removePunc(text)
    text = text.replace('ঃ', ' ')

    return text

def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
         "]+", re.UNICODE)
    return re.sub(emoj, ' ', data)

def lowercase_all_english_letters(text):
    pattern = re.compile(r"[A-Z]")
    lowercased_text = pattern.sub(lambda match: match.group().lower(), text)
    return lowercased_text

def preprocessing(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    txt = url_pattern.sub('', text)
    txt = remove_emojis(txt)
    txt = lowercase_all_english_letters(txt)
    return normalizeUnicode(txt)
```

```python
from datasets import Dataset,DatasetDict
import tensorflow as tf

train_ds = Dataset.from_pandas(train, preserve_index=False)
eval_ds = Dataset.from_pandas(test, preserve_index=False)

dataset = DatasetDict({
    "train": Dataset.from_pandas(train, preserve_index=False),
    "validation": Dataset.from_pandas(test, preserve_index=False)
    })
train_ds
```

```python
checkpoint = "csebuetnlp/banglabert"
# checkpoint = "csebuetnlp/banglabert_large"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

def preprocess_function(examples):
    return tokenizer(examples["summary"], truncation=True, padding=True, max_length=512)
```

```python
tokenized_data = dataset.map(preprocess_function, batched=True)


tokenized_datasets = train_ds.map(preprocess_function, batched=True)
tokenized_val_datasets = eval_ds.map(preprocess_function, batched=True)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
# tokenized_train_datasets = tokenized_datasets.remove_columns(["cleaned"])
tokenized_train_datasets = tokenized_datasets.rename_column("label", "labels")
tokenized_train_datasets.set_format("torch")
print(tokenized_train_datasets.column_names)

# tokenized_eval_datasets = tokenized_val_datasets.remove_columns(["cleaned"])
tokenized_eval_datasets = tokenized_val_datasets.rename_column("label", "labels")
tokenized_eval_datasets.set_format("torch")
print(tokenized_eval_datasets.column_names)
```

```python
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)

    # Calculate accuracy
    accuracy = accuracy_score(labels, preds)

   # Calculate precision, recall, and F1-score
    precision = precision_score(labels, preds, average='macro')
    recall = recall_score(labels, preds, average='macro')
    f1 = f1_score(labels, preds, average='macro')

    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }
```

```python
num_epochs = 4

training_args = TrainingArguments(output_dir='model_final/',
                                  evaluation_strategy="epoch",
                                  num_train_epochs=num_epochs,
                                  learning_rate = 5e-5,
                                  weight_decay = 5e-7,
                                  greater_is_better=False,
                                  optim='adafactor',
                                  save_strategy="no" 
                                 )

model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2,
                                           id2label = {0:"positve", 1:"negative"},
                                           label2id = {"positve":0, "negative":1}
)

trainer = Trainer(
    model,
    training_args,
    train_dataset=tokenized_train_datasets,
    eval_dataset=tokenized_eval_datasets,
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, auc

def calculate_classification_metrics(preds, true_values):
    """
    Calculates classification metrics and prepares data for visualization.

    Args:
        preds: A NumPy array of predicted class labels.
        true_values: A NumPy array of ground truth class labels.

    Returns:
        A dictionary containing:
            confusion_matrix: A 2D NumPy array representing the confusion matrix.
            roc_auc: The Area Under the ROC Curve (AUC) score.
            fpr: False Positive Rate for ROC curve.
            tpr: True Positive Rate for ROC curve.
    """

    # Ensure binary classification for ROC curve (can be adapted for multi-class)
    if len(np.unique(true_values)) > 2:
        raise ValueError("ROC curve currently supports binary classification only.")

    # Calculate confusion matrix
    cm = confusion_matrix(true_values, preds)

    # Calculate ROC curve metrics (assuming binary classification)
    fpr, tpr, thresholds = roc_curve(true_values, preds)
    roc_auc = auc(fpr, tpr)

    return {
        "confusion_matrix": cm,
        "roc_auc": roc_auc,
        "fpr": fpr,
        "tpr": tpr,
    }
```

```python
metrics = calculate_classification_metrics(preds, true)

# Confusion Matrix (Heatmap visualization recommended for large matrices)
print("Confusion Matrix:\n", metrics["confusion_matrix"])

# ROC Curve
plt.figure(figsize=(8, 6))
plt.plot(metrics["fpr"], metrics["tpr"], label='ROC curve (area = %0.2f)' % metrics["roc_auc"])
plt.plot([0, 1], [0, 1], 'k--', label='No Skill')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.savefig("ROC_sentiment_nodis.jpg")
plt.show()
```
